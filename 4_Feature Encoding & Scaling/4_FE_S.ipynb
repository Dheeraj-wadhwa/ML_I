## IMPORT LIBRARIES


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

## LOAD DATASET

```python
df = pd.read_csv("adult.csv")
df.head(5)
```

```python
df.info()
```

## Identify categorical and numerical features

```python
categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['int64']).columns
```

```python
list(categorical_cols)
```

```python
list(numerical_cols)
```

```python
df.isnull().sum()
```

## Apply Label Encoding where order exists

```python
label_enc_cols = ['income']
```

```python
le = LabelEncoder()
for col in label_enc_cols:
    df[col] = le.fit_transform(df[col])
```

```python
label_enc_cols
```

## Apply One-Hot Encoding where order does not exist

```python
onehot_cols = list(set(categorical_cols) - set(label_enc_cols))
df = pd.get_dummies(df, columns=onehot_cols, drop_first=True)
```

```python
onehot_cols
```

## Scale numerical features using StandardScaler

```python
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

## Compare model readiness before and after scaling

```python
df.head()
```

```python
df.shape
```

## Save processed dataset

```python
df.to_csv("adult_preprocessed.csv", index=False)
```

* Categorical features converted to numeric format
* Ordered category used Label Encoding
* Unordered categories used One-Hot Encoding
* Numerical features scaled to mean=0 and std=1
* Dataset is now ML-model ready
